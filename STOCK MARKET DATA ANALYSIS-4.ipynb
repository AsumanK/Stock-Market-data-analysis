# Resampling 
This involves changing the frequncy of your time series observations 
1. Down-sampling
This involves decreasing the frequency of the samples such annual into monthly.
2. Up-sampling
This involves increasing the frequency of the samples such as hours to minutes.

In this case, we will look at 2019, 7, 5 to 2020, 8, 5 for all stocks.

# Moving Windows within each stock market
Moving windows exists when you compute the statistics on a window of data represented by a specific period of time and then slide the window across the data by the specified time. That way, the statistic is continually calculated as long as the window falls first within the dates of time series.

The rolling mean smoothens out short-term fluctuations and highlights longer-term trends in data.

# Volatility of stocks

The volatility of a stock is the measurement of change in variance of a stock over a specified period of time.

The aim of comparing the volatility of a stock with another stock is to find out which stock is less riskier or comparing to the market index is to find out which stock is less riskier in the overall market.

The higher the volatility, the riskier the investment in that stock.

```python
#for bpl
min_periods = 75 

# Calculate the volatility
vol = daily_pct_change_bpl.rolling(min_periods).std() * np.sqrt(min_periods) 

vol.fillna(0,inplace=True)

vol.tail()
```

```python
# Plot the volatility
vol.plot(figsize=(10, 8))

# Show the plot
plt.show()
```

Ouput shows the stock was very volatile at specific periods in 2009 to 2019, then 2011, a period after the 2008 financial crisis and very volatile during the pandemic (covid 19).

```python
#for ulvrl
min_periods = 75 

# Calculate the volatility
vol1 = daily_pct_change_ulvrl.rolling(min_periods).std() * np.sqrt(min_periods) 

vol1.fillna(0,inplace=True)

vol1.tail()
```

```python
# Plot the volatility
vol1.plot(figsize=(10, 8))

# Show the plot
plt.show()
```

# Rolling Means (Trends and Seasonality)
Instead of variance, the stocks show us short and long term windows


```python
bpl_adj_close_px = bpl['Adj Close']
# Short moving window rolling mean (monthly)
bpl['42'] = bpl_adj_close_px.rolling(window=40).mean()

# Long moving window rolling mean (yearly)
bpl['252'] = bpl_adj_close_px.rolling(window=252).mean()

# Plot the adjusted closing price, the short and long windows of rolling means
bpl[['Adj Close', '42', '252']].plot(title="BP.L")

# Show plot
plt.show()

ulvrl_adj_close_px = ulvrl['Adj Close']
# Short moving window rolling mean
ulvrl['42'] = ulvrl_adj_close_px.rolling(window=40).mean()

# Long moving window rolling mean
ulvrl['252'] = ulvrl_adj_close_px.rolling(window=252).mean()

# Plot the adjusted closing price, the short and long windows of rolling means
ulvrl[['Adj Close', '42', '252']].plot(title="ULVR.L")

# Show plot
plt.show()

avl_adj_close_px = avl['Adj Close']
# Short moving window rolling mean
avl['42'] = avl_adj_close_px.rolling(window=40).mean()

# Long moving window rolling mean
avl['252'] = avl_adj_close_px.rolling(window=252).mean()

# Plot the adjusted closing price, the short and long windows of rolling means
avl[['Adj Close', '42', '252']].plot(title="AV.L")

# Show plot
plt.show()

ssel_adj_close_px = ssel['Adj Close']
# Short moving window rolling mean
ssel['42'] = ssel_adj_close_px.rolling(window=40).mean()

# Long moving window rolling mean
ssel['252'] = ssel_adj_close_px.rolling(window=252).mean()

# Plot the adjusted closing price, the short and long windows of rolling means
ssel[['Adj Close', '42', '252']].plot(title="SSE.L")

# Show plot
plt.show()
```

Outputs indicate the adj close, short term (trends) and long term (seasonality), however, its better to invest in long term when its stable compared to short term when its more volatile.
